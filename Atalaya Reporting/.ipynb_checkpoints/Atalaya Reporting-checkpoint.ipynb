```python
import FiOps_Functions as Funcs
import pandas as pd
import numpy as np
import time
import datetime
from datetime import datetime, timedelta
import sqlalchemy
import collections
```

```python
cnx = Funcs.sql_fiops_connect()
```

```python
query = """
SELECT DISTINCT
	CASE
    WHEN cases.LEASE_TRANSFER_PAYMENT_TYPE__C  = 'PP'
    THEN 'Pure Prepaid to Prepaid Transfer'
    WHEN cases.LEASE_TRANSFER_PAYMENT_TYPE__C  = 'MLP-PP'
    THEN 'Monthly to Prepaid Transfer'
    WHEN cases.LEASE_TRANSFER_PAYMENT_TYPE__C = 'MLP'
    THEN 'Pure Monthly to Monthly Transfer'
    END as 'Reassignment Res'
,	cases.CASENUMBER as 'Case Number'
,	cases.TYPE as 'Type'
,	M.ACCOUNT_NUMBER__C as 'Account Number'
,	M.Name as 'Project Name'
,	cast(M.CREATEDDATE as Date) as 'Project Created Date'
,	M.SALE_CLOSED__C as 'Sale Closed Date'
,	M.SITE_STATE_PROVINCE__C as 'State'
,	Util.Name as 'Utility'
,	t.Name as 'Tranche Name'
,	cases.LEASE_TRANSFER_PAYMENT_TYPE__C as 'Lease Transfer Payment Type'
,	cases.LEASE_TRANSFER_EFFECTIVE_DATE__C as 'Lease Transfer Effective Date'
,	M.FINAL_INTER_APPROVED__C as 'Final Inter Approved'
,	MONTH(M.FINAL_INTER_APPROVED__C) as 'PTO Month'
,	YEAR(M.FINAL_INTER_APPROVED__C) as 'PTO Year'
,	cast(cases.CREATEDDATE as Date) as 'Reassignment Created Date'
,	cast(cases.CLOSEDDATE as Date) as 'Reassignment Closed Date'
,	datediff(cases.CLOSEDDATE, cases.CREATEDDATE)/30 as 'Months Reassignment Open'
,	datediff(cases.LEASE_TRANSFER_EFFECTIVE_DATE__C, M.SALE_CLOSED__C) / 365 as 'Age of Customer Reassignment (Years)'
,	LAST_DAY(cases.CLOSEDDATE) as 'EO Reassignment Closed Month'
,	DATE_ADD(LAST_DAY(cases.CLOSEDDATE), INTERVAL 1 MONTH) as  'EO Reassignment Closed Month + 1'
,	DATE_ADD(M.FINAL_INTER_APPROVED__C, INTERVAL 1 MONTH) as 'Payment Start'
,	DATE_ADD(M.FINAL_INTER_APPROVED__C, INTERVAL ((12*SP_LSC.TERM__C)) MONTH) as 'Payment End'
,	SP_LSC.DOWN_PAYMENT__C as 'Down Payment'
,	SP_LSC.PAYMENT_PRE_TAX__C as 'Payment Pre Tax'
,	SP_LSC.ESCALATOR__C as 'Escalator'
,	SP_LSC.TERM__C as 'Term'


FROM
	salesforce_prod.Opportunity O
		LEFT JOIN salesforce_prod.Lead Lead ON O.ID = lead.CONVERTEDOPPORTUNITYID 
			AND Lead.Name NOT LIKE '%%testsungevity%%'
			AND Lead.Name NOT LIKE '%%sungevitytest%%'
			AND lead.`STATUS` NOT LIKE 'Qualified Duplicate'
		LEFT JOIN salesforce_prod.iQuote__c ON O.Id = iQuote__c.Opportunity__c
		LEFT JOIN salesforce_prod.Account Account ON iQuote__c.Account__c = Account.Id
        LEFT JOIN salesforce_prod.Milestone1_Project__c M ON M.Id = iQuote__c.Project__c 
        LEFT JOIN salesforce_prod.System__c System_SV4B ON iQuote__c.Id = System_SV4B.iQuote__c AND System_SV4B.SV4B__c = 1
        LEFT JOIN salesforce_prod.System__c System_LSC ON iQuote__c.Id = System_LSC.iQuote__c AND System_LSC.LSC__c = 1
        LEFT JOIN salesforce_prod.System__c System_LEPC ON iQuote__c.Id = System_LEPC.iQuote__c AND System_LEPC.LEPC__c = 1
        LEFT JOIN salesforce_prod.cases cases on cases.project__c = M.ID
        LEFT JOIN salesforce_prod.Tranche__c t ON (t.id = M.Tranche__c)
        LEFT JOIN salesforce_prod.utility__c Util on Util.ID = M.UTILITY__C
        LEFT JOIN salesforce_prod.AHJ__c ON M.AHJ_Name__c = AHJ__c.Id
        LEFT JOIN salesforce_prod.recordtype on recordtype.ID = cases.RECORDTYPEID
        LEFT JOIN salesforce_prod.Funding_Product__c FP ON (FP.Project__c = M.Id AND FP.Selected_Product__c = 1)
        LEFT JOIN salesforce_prod.Funding_Provider__c ON Funding_Provider__c.Id = FP.Funding_Provider__c   
        LEFT JOIN salesforce_prod.Funding_Provider__c final ON final.Id = FP.FINAL_ASSET_OWNER__C 
        LEFT JOIN salesforce_prod.jrs_project JRSP ON M.Id = JRSP.ProjectId
        LEFT JOIN salesforce_prod.jrs_funding_product_attribute FPA ON FP.Id = FPA.Funding_Product__c
        LEFT JOIN salesforce_prod.Region__c zip ON zip.Name = LEFT(iQuote__c.Postal_Code__c,5)
        LEFT JOIN salesforce_prod.utility_tarrifs__C tarrifs on tarrifs.ID = O.CURRENT_UTILITY_TARRIF__C
        LEFT JOIN salesforce_prod.System_Price__c SP_LSC ON (System_LSC.id=SP_LSC.System__c AND FP.Id = SP_LSC.Funding_Product__c AND SP_LSC.Active_Pricing__c = 1)
        
WHERE
(t.NAME LIKE '%%CITI%%' or t.Name LIKE '%%RABO%%' or t.Name LIKE '%%USB%%')
AND cases.LEASE_TRANSFER_STATUS__C = 'Escrow Closed'
AND (cases.LEASE_TRANSFER_EFFECTIVE_DATE__C <> '' and cases.LEASE_TRANSFER_EFFECTIVE_DATE__C IS NOT NULL)
AND recordtype.Name = 'Compliance'
AND (cases.Subject NOT LIKE '%%duplicate%%' or cases.Subject NOT LIKE '%%test%%' or cases.Subject NOT LIKE '%%dupe%%')
AND cases.TYPE = 'Lease Transfer'
AND cases.LEASE_TRANSFER_EFFECTIVE_DATE__C <= 20161231
"""
```

```python
df = pd.read_sql_query(query, cnx)
df['Lease Transfer Effective Date'] = pd.to_datetime(df['Lease Transfer Effective Date'])
```

```python
return_indicies = []

mlp_pp_i = df['Lease Transfer Payment Type'] == 'MLP-PP'
pp_i = df['Lease Transfer Payment Type'] == 'PP'
mlp_i = df['Lease Transfer Payment Type'] == 'MLP'


for project in unique_projects:
    
    project_i = df['Project Name'] == project
    mlp_pp_df = df[project_i & mlp_pp_i]
    pp_df = df[project_i & pp_i]
    mlp_df = df[project_i & mlp_i]
    
    if mlp_pp_df.shape[0] > 0:
        return_indicies.append(mlp_pp_df['Lease Transfer Effective Date'].idxmax())         
    
    elif pp_df.shape[0] > 0:
        return_indicies.append(pp_df['Lease Transfer Effective Date'].idxmin())
        
    elif mlp_df.shape[0] > 0:
        return_indicies.append(mlp_df['Lease Transfer Effective Date'].idxmin())
```
